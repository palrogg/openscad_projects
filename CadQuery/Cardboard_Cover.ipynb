# Cardboard Cover

Replacement for a Memory Cardboard Cover. Change dimensions to fit your needs.

```python
import cadquery as cq
from IPython.display import clear_output
```

```python
INNER_WIDTH = 158
INNER_LENGTH = 100
INNER_HEIGHT = 32
THICKNESS = 0.8
```

```python
clear_output(wait=True)
width = INNER_WIDTH + 2 * THICKNESS
length = INNER_LENGTH + 2 * THICKNESS
height = INNER_HEIGHT + THICKNESS

outer_box = cq.Workplane("front").box(width, length, height).fillet(0.5 * THICKNESS)
negative_part = cq.Workplane("front", origin=(0, 0, 0.5 + THICKNESS)).box(
    INNER_WIDTH, INNER_LENGTH, INNER_HEIGHT
)
result = outer_box.cut(negative_part)
display(result)
```

```python
cq.exporters.export(result, "output/cardboard_cover.step")
```
