# Furniture Leg Support

I needed to raise furniture legs a bit to protect them from water.

```python
import cadquery as cq
from IPython.display import clear_output
```

```python
clear_output(wait=True)
inner_width = 35
inner_length = 25
wall_thickness = 1.4
tolerance = 0.5

width = inner_width + 2 * (wall_thickness + tolerance)
length = inner_length + 2 * (wall_thickness + tolerance)
height = 5

outer_shape = cq.Workplane("front").box(width, length, height).fillet(0.4)
negative_part = cq.Workplane("front", origin=(0, 0, wall_thickness)).box(
    inner_width, inner_length, height
)
result = outer_shape.cut(negative_part)
display(result)
```

```python
cq.exporters.export(result, "output/furniture_leg_support.step")
```
