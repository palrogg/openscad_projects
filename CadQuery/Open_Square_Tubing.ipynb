# Open Seam Square Tubing

This was just what I needed for a shelf I built a few years back.

```python
import cadquery as cq
from IPython.display import clear_output
```

```python
clear_output(wait=True)
height = 100
square_tube = (
    cq.Workplane("front")
    .box(25.2, 25.2, height)
    .faces("|Z")
    .shell(1.2)
)
negative_part = cq.Workplane("front", origin=(10, 0, 0)).box(30, 23, height*1.2)
result = square_tube.cut(negative_part)
display(result)
```

```python
cq.exporters.export(result, "output/open_square_tubing.step")
```
