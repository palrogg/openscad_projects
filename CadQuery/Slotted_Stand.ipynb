# Slotted Stand

Stand for a 3.5mm thick wooden panel.

```python
import cadquery as cq
from IPython.display import clear_output
```

```python
test = False
clear_output(wait=True)
width = 30
height = 50
depth = 30

if test:
    width = 10
    height = 10
    depth = 20
plywood_thickness = 3.5
tolerance = 0.1
slot_thickness = plywood_thickness + tolerance

square_tube = cq.Workplane("front").box(width, depth, height).fillet(0.4)
negative_part = cq.Workplane("front", origin=(0, 0, 0.25 * height)).box(
    slot_thickness, 50, 0.5 * height
)
result = square_tube.cut(negative_part)
display(result)
```

```python
cq.exporters.export(result, "output/slotted_stand_test.step" if test else "output/slotted_stand.step")
```
